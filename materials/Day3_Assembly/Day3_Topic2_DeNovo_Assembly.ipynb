## De Novo Assembly Algorithms

### Building a Genome from Scratch

When you don't have a reference, you must overlap reads to build **Contigs**.

*   **De Bruijn Graphs**: Mathematical model using k-mers to find overlaps.
*   **Unicycler**: A modern hybrid assembler that excels at bacterial genomes.
