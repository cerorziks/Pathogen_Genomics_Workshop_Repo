# Day 3: Topic 1
## Read Mapping & Alignment

### Concept of Mapping

Comparing your sequenced fragments to a known **Reference Genome**.

*   **Tool**: BWA-MEM is the standard for short reads.
*   **BAM/SAM**: File formats that store where each read "stuck" to the reference.
*   **Coverage**: How many times each base in the reference was sequenced.
