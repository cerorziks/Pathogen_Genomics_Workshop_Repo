## Genome Annotation

### Finding the Genes

Assembly gives you letters; **Annotation** gives you function.

*   **Structural Annotation**: Where the genes are (Start/Stop codons).
*   **Functional Annotation**: What the genes do (BLASTing against databases like UniProt).
