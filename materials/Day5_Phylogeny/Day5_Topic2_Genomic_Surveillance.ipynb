## Genomic Surveillance & One Health

### Scaling Up to Public Health

Genomic surveillance is the continuous monitoring of changes in pathogens to inform health decisions.

*   **SNP-based Resolution**: Allows high-resolution outbreak detection.
*   **MLST**: Multi-Locus Sequence Typing. Standardized global nomenclature.
