## Phylogenetics Basics

### Understanding Evolution

**Phylogenetics** is the study of evolutionary relationships among biological entities.

*   **Phylogenetic Tree**: A diagram showing the inferred evolutionary relationships.
*   **Tips**: Individual isolates or entities being compared.
*   **Branches**: The length represents evolutionary distance/time.
*   **Nodes**: Points where lineages diverge (Common ancestors).

### Methods

*   **Neighbor-Joining**: Fast, distance-based method.
*   **Maximum Likelihood**: Statistical method that finds the tree that most likely explains the data.
*   **Bootstrapping**: A method for assessing the robustness of the tree branches.
