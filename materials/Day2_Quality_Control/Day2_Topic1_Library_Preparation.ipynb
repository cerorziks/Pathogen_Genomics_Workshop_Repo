## NGS Library Preparation

### What is a Library?

A "Library" is a collection of DNA fragments prepared for sequencing by adding specific adapters to their ends.

*   **Fragmentation**: DNA is broken into smaller pieces.
*   **Ligation**: Adapters (handles) are attached to both ends.
*   **Indexing**: Barcodes are added to identify different samples in one run.

### Steps in the Lab

1.  **Shearing**: Mechanical or enzymatic cutting.
2.  **End Repair**: Making ends blunt or adding an "A" overhang.
3.  **Size Selection**: Keeping only fragments of a specific length (e.g., 300-500bp).
4.  **Quantification**: Measuring how much library you actually have.
