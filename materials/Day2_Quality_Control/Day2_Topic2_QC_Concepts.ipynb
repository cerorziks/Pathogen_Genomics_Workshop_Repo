## Quality Control Concepts

### Why QC matters?

"Garbage In, Garbage Out". If your input data is low quality, your assembly or variant calling will be unreliable.

*   **Phred Score (Q)**: Measures base call accuracy.
    *   Q20 = 99% accuracy.
    *   Q30 = 99.9% accuracy (The standard).
    *   Q40 = 99.99% accuracy.

Follow these steps in Galaxy to clean your raw reads:

1.  **FastQC (Raw)**: Run on your raw R1 and R2 files. 
2.  **Inspect**: Check the 'Per base sequence quality'. Does it drop at the 3' end?
3.  **Trimming**: Run **Trimmomatic** or **Cutadapt**.
    *   *Setting*: Sliding window (4:20).
    *   *Setting*: Remove adapters (Illumina TruSeq/Nextera).
4.  **FastQC (Clean)**: Run again on the trimmed files to verify success.

## üõ†Ô∏è Practical 2: Nanopore QC

Long reads require different metrics than short reads.

1.  **NanoPlot**: Run on a provided Nanopore FASTQ file.
2.  **Inspect**: Look at the "Read Length vs Quality" scatter plot.
3.  **Metrics**: What is the **N50**? (The length at which 50% of the total bases are in reads of this length or longer).
