<!-- Truseq style PCR indexing primers

The IDT and other available UDI barcoding primers will have sequence:
P7
CAAGCAGAAGACGGCATACGAGAT[barcode]GTGACTGGAGTTCAGACGTGTGCTCTTCCGATC*T

P5 indexing oligo
AATGATACGGCGACCACCGAGATCTACAC[barcode]ACACTCTTTCCCTACACGACGCTCTTCCGATC*T


Your PCR primers will need Illumina flanking sequences at 5’ end compatible with these sequences so should have general design

Forward primer would have design
5’ ACACTCTTTCCCTACACGACGCTCTTCCGATCT gene specific sequence 3’

Reverse primer would have design
5’ GTGACTGGAGTTCAGACGTGTGCTCTTCCGATCT gene specific sequence 3’
Using these sequences means that custom sequencing primers are not required. -->
