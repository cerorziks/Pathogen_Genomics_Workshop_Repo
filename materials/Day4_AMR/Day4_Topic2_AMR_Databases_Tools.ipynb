# Day 4: Topic 2
## AMR Databases & Detection Tools

### The Big Databases

Predicting resistance relies on high-quality curated databases.

*   **CARD**: Comprehensive Antibiotic Resistance Database. Focuses on mechanisms.
*   **ResFinder**: Focuses on acquired resistance genes.
*   **NCBI AMRFinderPlus**: Highly curated, used by official public health labs.
