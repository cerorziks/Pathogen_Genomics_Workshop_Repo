## Antimicrobial Resistance (AMR): Mechanisms

### How Bacteria Survive Antibiotics

1.  **Deactivation**: Enzymes that destroy the drug (e.g., Beta-lactamases).
2.  **Efflux**: Pumping the drug out of the cell (e.g., TetA efflux pump).
3.  **Target Modification**: Changing the drug's target site so it no longer binds (e.g., mecA in MRSA).

### Acquisition of Resistance

*   **Mutational**: Small changes in core genes passed vertically.
*   **Acquired**: Horizontal Gene Transfer (HGT) via Plasmids, Transposons, or Integrons.
*   **Intrinsic**: Natural, inherent resistance of a species.
