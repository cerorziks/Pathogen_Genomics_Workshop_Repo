## Genomic Data Formats

### FASTA vs FASTQ

![File Format Differences](images/Formats-016.jpg)

*   **FASTA**: Header line (`>`) + Sequence. Used for references.
*   **FASTQ**: Header (`@`) + Sequence + Separator (`+`) + Quality scores. Used for raw reads.
