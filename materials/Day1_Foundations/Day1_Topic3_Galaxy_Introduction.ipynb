## Introduction to Galaxy & Cloud Computing

### What is Galaxy?

Galaxy is an open, web-based platform for accessible, reproducible, and transparent biomedical research.

*   **Accessible**: Users without programming experience can easily specify parameters and run tools and workflows.
*   **Reproducible**: Galaxy captures information so that any user can repeat and understand a complete computational analysis.
*   **Transparent**: Galaxy allows users to share and publish analyses via the web and create Pages.

### The Galaxy Interface

1.  **Tools Panel (Left)**: Search for and select analysis tools.
2.  **Middle Panel (Center)**: The workspace where you view datasets and run tools.
3.  **History Panel (Right)**: A chronological list of all your datasets and analysis steps.
