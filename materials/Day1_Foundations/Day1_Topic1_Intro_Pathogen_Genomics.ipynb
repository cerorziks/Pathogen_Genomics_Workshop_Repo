# Pathogen Genomics Workshop
## Topic 1: Introduction to Pathogen Genomics

### What is Pathogen Genomics?

*   **Genomics**: Reading the complete DNA instruction book (genome) of an organism.
*   **Pathogen Genomics**: Specifically applying this to bacteria, viruses, fungi, and parasites that cause disease.
*   **The Shift**: Moving from looking at *parts* of the bug (PCR, serology) to the *whole* bug (WGS).

### Why is it a Game Changer?

1.  **Universal Test**: One test (sequencing) can identify the species, drug resistance, and virulence factors.
2.  **Resolution**: Can distinguish between isolates that look identical by other methods.
3.  **Speed**: With technologies like Nanopore, we can go from sample to answer in hours.

### Public Health Applications

*   **Outbreak Detection**: Linking cases that are otherwise unconnected.
*   **Source Tracking**: Finding the contaminated water source or food product.
*   **AMR Surveillance**: Monitoring the spread of "superbugs" across borders.
*   **Vaccine Evaluation**: Checking if the pathogen has mutated to escape the vaccine.
